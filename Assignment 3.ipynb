Q1- Why are functions advantageous to have in your programs?
ANS-

1-Reusability: Functions allow you to write a block of code that can be used multiple times throughout your program without having to rewrite the code each time. This makes your code more efficient and easier to maintain.

2-Modularity: Functions allow you to break down a large program into smaller, more manageable modules. This makes your code easier to read and understand, and it also makes it easier to debug.

3-Abstraction: Functions provide a level of abstraction, which means that you can use a function without having to understand how it works internally. This allows you to focus on the high-level functionality of your program, rather than the details of how it is implemented.

4-Encapsulation: Functions can encapsulate code and data, which means that they can hide implementation details and prevent other parts of your program from accessing them. This improves security and reduces the risk of errors and bugs.

5-Testing: Functions make it easier to test your code, since you can test each function in isolation before integrating it into the larger program. This helps you to catch errors and bugs early on, when they are easier to fix.

Q2-When does the code in a function run: when its specified or when its called?
ANS-

The code in a function runs when it is called, not when it is specified.
When you define a function, you are essentially creating a block of code that performs a specific task or set of tasks and giving it a name that can be used to call the function later. The code inside the function is not executed at the time of definition. Instead, the code is stored in memory, and the function is registered as a callable object.

When the function is called, either by the main program or by another function, the code inside the function is executed. The function takes any input arguments that were passed to it, performs its designated task, and then returns any output values that were generated by the code.

Q4-What is the difference between a function and a function call?
ANS-

A function is a block of code that performs a specific task or set of tasks. It is defined using the def keyword in Python, and can take zero or more input parameters and return zero or more output values.

On the other hand, a function call is an instruction that tells the program to execute a specific function with a given set of input parameters. When a function is called, the program jumps to the function definition, executes the code block inside it, and then returns to the point in the program where the function was call

Q5-How many global scopes are there in a Python program? How many local scopes?
ANS-

There is only one global scope, which is created when the program starts executing. The global scope is the highest level of the program's namespace hierarchy, and any variables defined in this scope are accessible from any part of the program, including from within functions.

On the other hand, there can be multiple local scopes in a Python program. A local scope is created whenever a function is called, and it exists only for the duration of that function call. Any variables defined within a function are only accessible within that function and its nested functions. Once the function call completes, the local scope is destroyed, and any variables defined within it are lost.


Q6-What happens to variables in a local scope when the function call returns?
ANS-

When a function call returns, any variables defined within the local scope of that function are destroyed. This means that they are no longer accessible from any part of the program, and their values are lost.

Q14-What is the purpose of the try clause? What is the purpose of the except clause?
ANS-

The purpose of the try clause is to attempt to execute a block of code that might raise an exception. The code within the try block is executed until an exception is encountered or until the block of code completes without raising an exception. If an exception is raised, the code execution within the try block stops, and the control is transferred to the except block.

The purpose of the except clause is to handle the exception that was raised within the try block. The code within the except block is executed only if an exception occurs in the try block. You can use the except clause to catch the exception and take appropriate action, such as logging an error message, retrying the code block, or handling the exception in some other way.
