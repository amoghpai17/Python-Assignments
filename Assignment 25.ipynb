2)What is the relationship between generators and iterators?

ANS-
A generator is a special type of iterator that is defined using a function with a yield statement. When a generator function is called, it returns a generator object that can be used to iterate over the values produced by the function. Each time the yield statement is encountered, the current value is returned to the caller, and the state of the function is saved so that it can be resumed later to produce the next value.

In other words, generators are a way to create iterators using a simpler and more concise syntax, by using a function that can maintain its own state and return values one at a time, as needed.

3)What are the signs that a function is a generator function?

ANS-
1 - The function uses the yield keyword at least once to produce a value.

2 - When called, the function returns a generator object, which can be used to iterate over the sequence of values produced by the function.

3 - The function's execution is suspended and resumed each time a value is generated by the yield statement, allowing the function to maintain its state across multiple iterations.

4 - The function can be used in a for loop or with any other iterable function or method, just like any other iterator.

4)What is the purpose of a yield statement?

ANS-
The yield statement is used in a generator function to produce a sequence of values that can be iterated over. The purpose of the yield statement is to temporarily suspend the execution of the function and return a value to the caller, while preserving the function's state. When the function is called again, execution resumes from the point where it was suspended, and continues until the next yield statement is encountered.
