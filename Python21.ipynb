Q1. What is a probability distribution, exactly? If the values are meant to be random, how can you
predict them at all?

ANS-
A probability distribution is a mathematical function that describes the likelihood of different outcomes or events in a random process. It provides a way to represent the probability of each possible outcome or value of a random variable.

When we say that the values in a probability distribution are "random," we mean that there is inherent uncertainty or variability in the process being modeled. However, just because an outcome is random does not mean that we cannot make predictions about it. In fact, probability theory provides a framework for making predictions about the likelihood of different outcomes, even in situations where individual outcomes are uncertain.

Q2. Is there a distinction between true random numbers and pseudo-random numbers, if there is
one? Why are the latter considered “good enough”?

ANS-
Yes, there is a distinction between true random numbers and pseudo-random numbers.

True random numbers are generated from a physical process that is inherently random, such as atmospheric noise, radioactive decay, or thermal noise. Because these processes are truly random, the resulting numbers are also truly random and cannot be predicted or reproduced.

In contrast, pseudo-random numbers are generated by algorithms that produce a sequence of numbers that appear to be random, but are actually deterministic and reproducible. These algorithms typically start with a seed value and use mathematical formulas to generate a sequence of numbers that have statistical properties similar to those of truly random numbers. However, the sequence is not truly random and can be predicted if the seed value and algorithm are known.

Pseudo-random numbers are considered "good enough" for many applications because they have statistical properties that make them suitable for use in simulations, modeling, and other computational tasks where true randomness is not required. 

Q3. What are the two main factors that influence the behaviour of a 'normal' probability distribution?

ANS-
The behavior of a normal probability distribution is primarily influenced by its mean and standard deviation.

The mean, also known as the expected value, is the central point of the distribution and represents the average value of the data. It determines the location of the peak of the distribution. If the mean changes, the entire distribution will shift to the left or right.

Q4. Provide a real-life example of a normal distribution.

ANS-
 example of a real-life normal distribution is the distribution of heights of adult humans. In a large population of adult humans, the heights of individuals will naturally vary, but the distribution of heights will follow a normal distribution pattern.

In this distribution, the mean (average height) will be the central point, and the standard deviation will determine how spread out or tightly clustered the heights are around the mean. For example, the mean height for adult males in the United States is approximately 5 feet 9 inches, with a standard deviation of about 3 inches. This means that the majority of adult males will fall within a few inches of the mean height, with fewer individuals being significantly taller or shorter.

The normal distribution of height is an example of a continuous variable that follows a bell-shaped curve. It is used in many fields, such as medical research, to study the characteristics of human populations and determine the prevalence of certain conditions or traits among different groups.

Q5. In the short term, how can you expect a probability distribution to behave? What do you think will
happen as the number of trials grows?

ANS-
In the short term, a probability distribution can behave unpredictably due to random variations and fluctuations. However, as the number of trials or observations increases, the distribution will tend to converge towards a more stable and predictable pattern.

For example, if we flip a coin a small number of times, the distribution of outcomes may not appear to follow a clear pattern, and we may observe more variability in the results. However, as we increase the number of coin flips, we would expect the distribution of outcomes to more closely resemble a normal distribution, with an equal probability of getting heads or tails, and fewer extreme values.

This is due to the law of large numbers, which states that as the sample size increases, the sample mean will tend to approach the true population mean. As a result, the distribution of outcomes becomes more stable and predictable as the number of trials or observations grows.
